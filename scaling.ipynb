```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
boston_dataset = load_boston()
data = pd.DataFrame(boston_dataset.data,
columns=boston_dataset.feature_names)
data['MEDV'] = boston_dataset.target
```

```python
print(data.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 data.drop('MEDV', axis=1), data['MEDV'], test_size=0.3,
random_state=0)
```

# STANDARDIZATION/Z SCORE 

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,
columns=X_train.columns)
```

```python
scaler.mean_
```

```python
scaler.scale_
```

```python
# violin plot using scaled  train set values
sns.violinplot(data=X_train_scaled, orient='h')
plt.title('Violin Plot of Scaled Features')
plt.xlabel('Scaled Values')
plt.ylabel('Features')
plt.show()
```

```python
#visualization of the non scaled features using violin plot (train set m)
sns.violinplot(data=X_train, orient='h')
plt.title('Violin Plot of Scaled Features')
plt.xlabel('Feature Values')
plt.ylabel('Features')
plt.show()
```

# VECTOR UNIT LENGTH  SCALING

```python
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
```

```python
boston_dataset = load_boston()
data = pd.DataFrame(boston_dataset.data,
columns=boston_dataset.feature_names)
data['MEDV'] = boston_dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 data.drop('MEDV', axis=1), data['MEDV'], test_size=0.3,
random_state=0)
```

```python
scaler = Normalizer(norm='l1')
```

```python
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
np.round( np.linalg.norm(X_train_scaled, ord=1, axis=1), 1)
```
